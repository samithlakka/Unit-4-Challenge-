**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great!

```python
# Displays the first five rows of the cleaned DataFrame for inspection
print(properties_clean.head())
```

```python
 # Removes leading and trailing whitespace from all column names in the DataFrame to prevent potential issues
properties_clean.columns = properties_clean.columns.str.strip()
```

```python
# Filters the DataFrame to display rows where 'London_Borough' is 'Barking & Dagenham' for detailed inspection
print(properties_clean[properties_clean['London_Borough'] == 'Barking & Dagenham'])
```

```python
def create_price_ratio(d):
    # Check if 'Year' column exists before accessing it
    if 'Year' in d.columns:
        try:
            # Get the average prices for 1998 and 2018
            y1998 = float(d['Average_Price'][d['Year'] == 1998].values[0])
            y2018 = float(d['Average_Price'][d['Year'] == 2018].values[0])

            # Debug print statements to check the values
            print(f"Borough: {d['London_Borough'].unique()}, 1998 Price: {y1998}, 2018 Price: {y2018}")

            # Calculate the ratio
            ratio = y2018 / y1998

            return ratio
        except Exception as e:
            # Handle any errors that occur during calculation
            print(f"Error for Borough: {d['London_Borough'].unique()} - {e}")
            return None
    else:
        # Handle the case where 'Year' column is missing
        print(f"'Year' column missing in data for Borough: {d['London_Borough'].unique()}")
        return None

# Convert the 'Month' column to datetime format
properties_clean['Month'] = pd.to_datetime(properties_clean['Month'])

# Create the 'Year' column
properties_clean['Year'] = properties_clean['Month'].dt.year

# Apply the function to each borough and store the results in a dictionary
diff = {}
for b in properties_clean['London_Borough'].unique():
    borough = properties_clean[properties_clean['London_Borough'] == b]
    result = create_price_ratio(borough)
    if result is None:
        result = 0  # Set to 0 if ratio calculation failed
    diff[b] = result

# Convert the dictionary to a DataFrame
ratios = pd.DataFrame(diff, index=[0])
print(ratios.head())  # Display the first few rows of the resulting DataFrame
```

```python
# Transposes the DataFrame, switching rows and columns
ratios_T = ratios.T
# Resets the index of the transposed DataFrame, converting the index into a column
ratios = ratios_T.reset_index()
ratios.head()
```

```python
# Renames the columns: 'index' to 'Borough' and '0' to '2018' for clearer, more descriptive column names
ratios.rename(columns={'index':'Borough', 0:'2018'}, inplace=True)
ratios.head()
```

```python
# Sorts the DataFrame by the '2018' column in descending order and selects the top 15 rows with the highest values
top15 = ratios.sort_values(by='2018',ascending=False).head(15)
print(top15)
```

```python
# Creates a bar plot with boroughs on the x-axis and 2018 values on the y-axis
ax = top15[['Borough','2018']].plot(kind='bar')
# Sets the x-axis tick labels to the names of the boroughs
ax.set_xticklabels(top15.Borough)
```
